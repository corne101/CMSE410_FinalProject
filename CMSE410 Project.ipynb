```python
#function to compute tissue specificity for each gene
#takes in each row from RPKM dataset and finds a Tau score
#outputs a Tau for each gene
def Tau(row):
    Tau = 0
    if np.min(row) >= 0:
        if np.max(row) != 0:
            x = 1-(row/np.max(row))
            summation = sum(x)
            Tau = summation/(len(row)-1)
        else:
            Tau = 0
    return Tau
```

```python
# function to create new dataset of Tau values
# uses Tau to get tissue specificity for each gene 
# takes gene identities that were split off from earlier datasets
# creates new dataframe of gene identities and Tau values
def Tau_df(df, genes):
    new_df = pd.DataFrame()
    new_df["gene_id"] = genes
    Tau_list = []
    for index, row in df.iterrows():
        T = Tau(row)
        Tau_list.append(T)
    new_df["Tau"] = Tau_list
    return new_df
        
```

```python
#condenses data to only include genes which exist in the linkage dataset
def condense_data(df, linkage, col):
    return df[df["gene_id"].isin(linkage[col]) == True]
```

```python
#function to find gene matches between Tau dataframe and the linkage dataset
#adds Tau scores to likage dataset for later plotting
def find_ortho_genes(df, linkage, gene_col_name, new_col_name):
    match_Tau = []
    match_gene = []
    df_genes = list(df["gene_id"])
    link_genes = list(linkage[gene_col_name])
    df_Tau = list(df["Tau"])
    for gene in link_genes:
        #print(gene)
        match = 0
        for index in range(len(df_genes) -1):
            if gene == df_genes[index] and match != 1:
                match_Tau.append(df_Tau[index])
                match_gene.append(df_genes[index])
                match = 1
        if match == 0:
            match_Tau.append("NaN")
            match_gene.append("NaN")
    print(len(link_genes), len(match_Tau))
    linkage[new_col_name] = match_Tau
    #linkage["gene match"] = match_gene
    return linkage
```

```python
# find pearson correlation coeff for each pair of y values
def find_r(actual, predict):
    actual = list(actual)
    predict = list(predict)
    corr_matrix = np.corrcoef(actual, predict)
    corr = corr_matrix[0,1]
    return corr
```
